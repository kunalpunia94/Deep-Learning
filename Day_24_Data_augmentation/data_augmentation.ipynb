```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img(r"D:\Deep Learning 100 days\Day_24_Data_augmentation\train\dog.jpg",target_size=(200,200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
#got from this website ->>https://blog.keras.io/building-powerful-image-classification-models-using-very-little-data.html
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
# fill_mode= many types just go through
```

```python
img = image.img_to_array(img) 
```

```python
type(img)
```

```python
img.shape
```

```python
input_size = img.reshape(1,200,200,3) #1 bcoz we are processing one one image
```

```python
#using 'flow' bcoz doing on  single image
i = 0
for batch in datagen.flow(input_size, batch_size=1,save_to_dir=r'D:\Deep Learning 100 days\Day_24_Data_augmentation\new_img',save_format='jpg'):
    i += 1
    if i == 15:
        break 
```
